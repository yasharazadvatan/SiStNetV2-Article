```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# CSV dosyasını oku
df1 = pd.read_csv('./run-output-data/class-accuracy-train.csv')
df2 = pd.read_csv('./run-output-data/no-obj-acc.csv')
 
# Çizgi grafiği oluştur
plt.plot(df1['Step'], df1['Value'], label='Class Accuracy/Train', color='blue')
plt.plot(df2['Step'], df2['Value'], label='No Obj. Accuracy/Train', color='red')


# Grafik başlığı ve eksen etiketleri ekleyebilirsiniz
# plt.title('İki Farklı Veri Seti')
plt.xlabel('Train')
plt.ylabel('Accuracy')

# Grafiği göster
plt.show()
```

```python
# İki farklı CSV dosyasını oku
df1 = pd.read_csv('./run-output-data/class-accuracy-train.csv')
df2 = pd.read_csv('./run-output-data/no-obj-acc.csv')
df3 = pd.read_csv('./run-output-data/obj-acc.csv')
df4 = pd.read_csv('./run-output-data/mAP.csv')
 
# İki ayrı plot oluşturun
fig, axs = plt.subplots(2, 2, figsize=(10, 8))  # 2 satır, 2 sütunlu subplot

# İlk plot için
axs[0, 0].plot(df1['Step'], df1['Value'], label='Class Accuracy/Train', linewidth=2, color='blue')
axs[0, 0].set_title('(a) Class Accuracy/Train')
axs[0, 0].set_xlabel('Train')
axs[0, 0].set_ylabel('Accuracy') 

# İkinci plot için
axs[0, 1].plot(df2['Step'], df2['Value'], label='No Obj. Accuracy/Train', linewidth=2, color='green')
axs[0, 1].set_title('(b) No Obj. Accuracy/Train')
axs[0, 1].set_xlabel('Train')
axs[0, 1].set_ylabel('Accuracy') 

# 3. plot için
axs[1, 0].plot(df3['Step'], df3['Value'], label='Obj. Accuracy/Train', linewidth=2, color='purple')
axs[1, 0].set_title('(c) Obj. Accuracy/Train')
axs[1, 0].set_xlabel('Train')
axs[1, 0].set_ylabel('Accuracy') 

# 4. plot için
axs[1, 1].plot(df4['Step'], df4['Value'], label='mAP/Train', linewidth=2, color='red')
axs[1, 1].set_title('(d) mAP/Train')
axs[1, 1].set_xlabel('Train')
axs[1, 1].set_ylabel('mAP') 

# Alt grafikler arasındaki boşluğu ayarla
plt.tight_layout()

# Grafiği göster
plt.show()
```

```python
df = pd.read_csv('./run-output-data/loss.csv')
 
# Çizgi grafiği oluştur
plt.plot(df['Step'], df['Value'], label='Loss/Train', linewidth=2, color='blue')

# Grafik başlığı ve eksen etiketleri ekleyebilirsiniz
plt.title('Loss/Train')
plt.xlabel('Train')
plt.ylabel('Train')
plt.legend()
# Grafiği göster
plt.show()
```

```python
# CSV dosyasını oku
df_car = pd.read_csv('./run-output-data/map/map-car.csv')
df_cyclist = pd.read_csv('./run-output-data/map/map-cyclist.csv')
df_dontcare = pd.read_csv('./run-output-data/map/map-dontcare.csv')
df_misc = pd.read_csv('./run-output-data/map/map-misc.csv')
df_pedestrian = pd.read_csv('./run-output-data/map/map-pedestrian.csv')
df_person_sitting = pd.read_csv('./run-output-data/map/map-person-sitting.csv')
df_tram = pd.read_csv('./run-output-data/map/map-tram.csv')
df_truck = pd.read_csv('./run-output-data/map/map-truck.csv')
df_van = pd.read_csv('./run-output-data/map/map-van.csv')
df_all = pd.read_csv('./run-output-data/mAP.csv')

fig, axs = plt.subplots(1, 1, figsize=(7, 10))

# Çizgi grafiği oluştur
plt.plot(df_car['Step'], df_car['Value'], label='Car', linewidth=2, color='blue')
plt.plot(df_cyclist['Step'], df_cyclist['Value'], label='Cyclist', linewidth=2, color='green')
plt.plot(df_dontcare['Step'], df_dontcare['Value'], label='Don\'t Care', linewidth=2, color='pink')
plt.plot(df_misc['Step'], df_misc['Value'], label='Misc.', linewidth=2, color='purple')
plt.plot(df_pedestrian['Step'], df_pedestrian['Value'], label='Pedestrian', linewidth=2, color='orange')
plt.plot(df_person_sitting['Step'], df_person_sitting['Value'], label='Person Sitting', linewidth=2, color='brown')
plt.plot(df_tram['Step'], df_tram['Value'], label='Tram', linewidth=2, color='cyan')
plt.plot(df_truck['Step'], df_truck['Value'], label='Truck', linewidth=2, color='black')
plt.plot(df_van['Step'], df_van['Value'], label='Van', linewidth=2, color='gray')
plt.plot(df_all['Step'], df_all['Value'], label='mAP', linewidth=2, color='red')
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.01))


# Grafik başlığı ve eksen etiketleri ekleyebilirsiniz
plt.title('Mean Average Precision(mAP)')
plt.xlabel('Train')
plt.ylabel('mAP')

# Grafiği göster
plt.show()
```
