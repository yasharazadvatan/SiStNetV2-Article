```python
import os

def WriteNewFile(labels, label_path, new_label_path='./new_path'):
    for label in labels:

        with open(label, 'r') as f:
            parts = label.split('/')
            file_name = parts[-1]

            label_lines = f.readlines()

            for label_line in label_lines:
                split_line = label_line.split(' ')

                if split_line[0] != '8':

                    dir_path = os.path.join(new_label_path, *parts[1:-1])
                    os.makedirs(dir_path, exist_ok=True)

                    new_path = os.path.join(dir_path, file_name)

                    with open(new_path, 'a') as new_file:
                        new_file.write(label_line)
```

```python
WriteNewFile(v_labels,v_labels_path)
```

```python
WriteNewFile(t_labels,t_labels_path)
```

```python
WriteNewFile(ve_labels,ve_labels_path)
```

```python
WriteNewFile(te_labels,te_labels_path)
```
